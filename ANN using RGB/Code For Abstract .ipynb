### Sigmoid Activation Function 

- Range = 0 to 1
- Derivative Range = 0 to 0.25
- Formula = 1/1 + e-x

```python
plt.figure(figsize = (10,5))
x = np.arange(-10, 10, 0.1)
s = 1 / (1 + np.exp(-x))
plt.plot(x, s)
sns.set_style("whitegrid")
plt.xlabel("X_values")
plt.title("Sigmoid activation function")
plt.ylabel("Range of sigmoid from 0 to 1")
plt.show()
```

### Relu Activation Function 

- Formula  = max(0,x) {x if x > 0
                  0 if x < 0}
- Range    = 0 to Infinity 

```python
def relu(x):
    return max(0.0, x)
```

```python
x = 1.0
print('Relu Activation function on %f gives =  %.1f' % (x, relu(x)))  # precision and width concepts for accepting only required values after . 
x = -10.0
print('Relu Activation function on %f gives = %.1f' % (x, relu(x)))
x = 0.0
print('Relu Activation function on %f  gives = %.1f' % (x, relu(x)))
x = 15.0
print('Relu Activation function on %f  gives = %.1f' % (x, relu(x)))
x = -20.0
print('Relu Activation function on %f  gives = %.1f' % (x, relu(x)))
```

### Tanh Activation Function 

- Formula = e**x - e**-x / e**x + e**-x
- Range = -1 to 1 

```python
plt.figure(figsize = (10,5))
x = np.arange(-10, 10, 0.1)
t = np.tanh(x)
plt.plot(x, t)
plt.title("tanh Activation Function")
plt.xlabel("x_values")
plt.ylabel("range of -1 to 1")
plt.show()
```

### Softmax Activation Function 

- Formula = F(xi) = e**xi / sum(e**xi) where i = 0,1,2,3...
- Range   = 0 to 1


```python
import numpy as np

def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum()

```

### Implementing Random weights 

- For developing ANN architecure we need to find_out Input layers of an image and hidden layers and output layers . But Although we have this architecture manually we need to implement weights for our neural networks . so For the better outcomes we are implementing random weights using Explicit constructrs . 

```python
class kamal_ann_art:
    def __init__(self, layers_dims, activations_fnc,  type_="classif"):
        self.type_ = type_
        self.W = [None] * (len(layers_dims)-1)
        self.b = [None] * (len(layers_dims)-1)
        self.out = [None] * (len(layers_dims)-1)
        self.layers_dims = layers_dims
        self.activations_fnc = activations_fnc
        for i in range(len(layers_dims)-1):
            self.b[i] = np.random.randn(layers_dims[i+1]).reshape(1, layers_dims[i+1])
            self.W[i] = np.random.randn(layers_dims[i], layers_dims[i+1])
    
    def sigmoid(self, x):
        return 1.0 / (1.0 + np.exp(-x))
        
    def tanh(self, x):
        return np.tanh(x)
        
    def relu(self, x):
        return np.maximum(x, 0)
    
    def softmax(self, x):
        exp_scores = np.exp(x)
        out = exp_scores / np.sum(exp_scores, axis=1, keepdims=True)
        return out
        
    def activation(self, x, t):
        t = t.lower()
        if t == "sigmoid":
            return self.sigmoid(x)
        elif t == "tanh":
            return self.tanh(x)
        elif t == "relu":
            return self.relu(x)
        elif t == "softmax":
            return self.softmax(x)
        else:
            print("Non-supported activation function {}".format(t))
            
    def multiply(self, x, W):
        m = np.dot(x, W)
        return m
    
    def add(self, m, b):
        return m + b
            
    def forward(self, x):
        input_ = x
        for i, activation in enumerate(self.activations_fnc) :
            weighted_sum = self.multiply(x=input_, W=self.W[i])
            weighted_sum = self.add(m=weighted_sum, b=self.b[i])
            input_ = self.out[i] = self.activation(x=weighted_sum, t=activation)
```

```python
def calling_networks(n_depth, n_size, activation, colormode, alpha):
    layers = [5] 
    for i in range(n_depth):
        layers.append(n_size)
    
    colormode = colormode.lower()
    
    # Finding best output layers based on color_mode we selected 
    
    if colormode in ["rgb"] : 
        if not alpha:
            layers.append(3)
        else:
            layers.append(4)
    else:
        print("Inserted colormode '{}' is not part ob supported ones: [rgb,hsv]".format(colormode))
    
    possibles = ["sigmoid", "tanh", "relu","softmax"]
    if not activation.lower() in possibles:
        print('defined activation {} not supported in {}'.format(activation, str(possibles)))
        return None
        
    activations_fnc = [activation] * (len(layers)-2)
    activations_fnc.append("sigmoid")
    
    return layers, activations_fnc   
```

```python
def activate_colors(nnet, x, y, r,  z1, z2, colormode, alpha):
    input_ = np.array([x, y, r, z1, z2], dtype=np.float32).reshape(1, 5)
    nnet.forward(input_)
    
    colormode = colormode.lower()
    
    if colormode == "rgb":
        r = nnet.out[len(nnet.out)-1][0][0]
        g = nnet.out[len(nnet.out)-1][0][1]
        b = nnet.out[len(nnet.out)-1][0][2]
        a_index = 3
    else:
        print("Inserted colormode '{}' will support only: [rgb]".format(colormode))
    if alpha: 
        a = 1-abs(2*nnet.out[len(nnet.out)-1][0][a_index]-1)
        a = 0.25 + 0.75*a
    else:
        a = 1.0
    
    return r, g, b, a
```
